### pandas排序
注意：
* `inplace`属性：
    * 当`inplace=True`时，是在原位置进行排序
    * 当`inplace=False`时，是新扩增了一列记录排序结果
* `ascending`属性：
    * 默认为True
    * 当设置`ascending=False`时，会进行降序排列

```python
# 默认按照从小到大的升序方式排列
food_info.sort_values("Sodium_(mg)", inplace=True)
# print(food_info["Sodium_(mg)"])

# 降序排列
food_info.sort_values("Sodium_(mg)", inplace=True, ascending=False)
print(food_info["Sodium_(mg)"])
```

## 泰坦尼克号船员获救数据

```python
import pandas as pd
import numpy as np

titanic_survival = pd.read_csv("titanic_train.csv")
titanic_survival.head()
```

### 数据缺失值寻找

#### 用切片法取年龄

```python
age = titanic_survival["Age"]
print(age.loc[0:10])
```

#### 判断是否为缺失值

```python
age_is_null = pd.isnull(age)
print(age_is_null)
```

#### 找出age中的缺失值及其序号

```python
age_null_true = age[age_is_null]
print(age_null_true)
```

#### 找出age缺失值的个数

```python
age_null_count = len(age_null_true)
print(age_null_count)
```

### 数据缺失值处理
* 为什么处理？因为不处理的话统计平均值会显示NAN
* 处理缺失值方法（很少直接去掉）  
  为保持其为完整样本，一般用以下值进行填充：
    * 年龄均值
    * 年龄中位数
    * 年龄众数

```python
mean_age = sum(titanic_survival["Age"]) / len(titanic_survival["Age"])
print(mean_age)

# 在求和时会出现问题
print(len(titanic_survival["Age"]))
print(sum(titanic_survival["Age"]))
```

#### 填充缺失值
`fill_na`

#### 去掉缺失值
`drop_na`：如果当前列(Age,Sex)有缺失值，那我直接把有缺失值的这一行样本去掉

```python
# 如果某行有缺失值，就将有缺失值的字段/列去掉
drop_na_columns = titanic_survival.dropna(axis=1)
# print(drop_na_columns)

# 检查如果某个字段有缺失值，则将缺失值所在行去掉
new_titanic_survival = titanic_survival.dropna(axis=0, subset=["Age", "Sex"])
print(new_titanic_survival)
```

#### 去掉缺失值求平均

##### 1.手动处理求平均

```python
# 利用索引取出非缺失值项
good_ages = titanic_survival["Age"][age_is_null == False]

# 将筛选余下的项目求平均
mean_age_afterdel = sum(good_ages) / len(good_ages)

# 打印均值
print(mean_age_afterdel)
```

##### 2.pandas求平均
pandas会自动去掉缺失值以取平均

```python
mean_age_pandas = titanic_survival["Age"].mean()
print(mean_age_pandas)
```

### Task：求1～3等仓船票价格

```python
# 试验田
print(titanic_survival[titanic_survival["Pclass"] == 1])
```

##### 1.手动统计船舱对应的票价均值

```python
passenger_classes = [1, 2, 3]
# 构造一个字典，以求船舱对应等船票价格
fares_by_class = {}

for this_class in passenger_classes:
    # 取出相应仓位的整体数据
    pclass_rows = titanic_survival[titanic_survival["Pclass"] == this_class]

    # 取出相应仓位的价格数据，定义到Fare列
    pclass_fares = pclass_rows["Fare"]

    # 求仓位的平均价格
    fare_for_class = pclass_fares.mean()

    # 将仓位和平均价格写成 key:value 的形式，存入字典fares_by_class = {}
    fares_by_class[this_class] = fare_for_class

print(fares_by_class)        
```

##### 2.pandas自动统计船舱对应的票价均值
`pivot_table`：数据透视表，统计多个量之间关系的函数
* 查看各个仓位平均获救多少人

```python
passenger_survival = titanic_survival.pivot_table(index="Pclass", values="Survived", aggfunc=np.mean)
print(passenger_survival)
```

* 查看各个仓位平均票价

```python
average_price = titanic_survival.pivot_table(index="Pclass", values="Fare", aggfunc=np.mean)
print(average_price)
```

* 统计各个船舱乘客的平均年龄

```python
# aggfunc是统计功能函数，如果不指明，默认的参数是np.mean，当然也可以设置为np.sum
passenger_age = titanic_survival.pivot_table(index="Pclass", values="Age")
print(passenger_age)
```

```python
# print(help(pandas.pivot_table))
```

* 查看登船**码头地点**与**总票价、总生存人数**的关系

```python
port_stats = titanic_survival.pivot_table(index="Embarked", values=["Fare","Survived"], aggfunc=np.sum)
print(port_stats)
```

### pandas自定义&调用函数
* 自定义函数的时候可以使用pandas的属性，如`loc[n]`
* 调用函数的时候使用`pandas_item.apply(pandas_function)`

#### 调用前100个值

```python
def hundredth_row(column):
    hundredth_item = column.loc[99]
    return hundredth_item

hundredth_row = titanic_survival.apply(hundredth_row)
print(hundredth_row)
```

#### 计算每个字段缺失值的个数

```python
def not_null_count(column):
    column_null = pd.isnull(column)
    null = column[column_null]
    return len(null)

column_null_count = titanic_survival.apply(not_null_count)
print(column_null_count)
```

#### 将每个数据的船舱等级数字变为等级名

```python
def which_class(row):
    pclass = row["Pclass"]
    if pd.isnull(pclass):
        return "Unknown"
    elif pclass == 1:
        return "First Class"
    elif pclass == 2:
        return "Second Class"
    elif pclass == 3:
        return "Third Class"

classes = titanic_survival.apply(which_class, axis=1)
print(classes)
```

#### 将Survived的数据01转换成状态字符

```python
# print(titanic_survival[["Survived", "Pclass"]])
```

```python
def which_class(row):
    # 取出PassengerId、Survived数据
    survived = row["Survived"]
    if pd.isnull(survived):
        return "Unknown"
    elif survived == 0:
        return "Dead"
    elif survived == 1:
        return "Alive"

classes = titanic_survival.apply(which_class, axis=1)
print(classes)
```

#### 连续值离散化——生成年龄标签
判断乘客是成年人还是未成年人

```python
def generate_age_label(row):
    age = row["Age"]
    if pd.isnull(age):
        return "unknown"
    elif age < 18:
        return "minor"
    else:
        return "adult"

age_labels = titanic_survival.apply(generate_age_label, axis=1)
print(age_labels)
```

* 统计乘客获救与年龄的关系

```python
titanic_survival["age_labels"] = age_labels
age_group_survival = titanic_survival.pivot_table(index="age_labels", values="Survived")
print(age_group_survival)
```

# Series结构

```python
import pandas as pd
```

## 基本认识
* DataFrame
    * Series
        * ndarry
        
### 查看Series类型

```python
fandango = pd.read_csv('fandango_score_comparison.csv')
series_film = fandango['FILM']
print(type(series_film))

series_rt = fandango['RottenTomatoes']

```

### 提取Series一列的一个片段

```python
print(series_film[0:5])
```

## 定义一个Series

```python
from pandas import Series

# 从pandas中取两个字段的值出来
film_names = series_film.values
print(type(film_names))

rt_scores = series_rt.values

# 定义Series：以film_names为指针，rt_scores为内容
series_custom = Series(rt_scores, index=film_names)

# 显示Series中指定想看的两个电影的评分：用字符串当索引（之前当索引是数值）
series_custom[['Minions (2015)', 'Leviathan (2014)']]

print('\n')
# 查看Series中当一个片段
five_ten = series_custom[5:10]
print(five_ten)
```

## Series的排序
### 对Series进行排序
很少使用：因为很少对Series直接处理

```python
original_index = series_custom.index.tolist()

# 查看tolist()后的类型
print(type(series_custom))
print(type(original_index))

print('\n')
# 查看Series的索引
# print(series_custom.index)

print('\n')
# 查看tolist()后的数据
# print(original_index)

print('\n')
# 将original_index重新排序，作为索引（转换成list再使用sorted()重新排序）
sorted_index = sorted(original_index)
# print(sorted_index)

print('\n')
# 将上面重新排序的索引，作为series_custom排序的依据（使用.reindex()重新排序）
sorted_by_index = series_custom.reindex(sorted_index)
print(sorted_by_index)
```

### 对Series按照键/值进行排序

```python
sc2 = series_custom.sort_index()
sc3 = series_custom.sort_values()

# 按照键排序
print(sc2[0:10])

# 按照值大小排序
print('\n')
print(sc3[0:10])
```

## Series的操作

两个维度相同的Series可以直接对应位置相加  
维度不相同，则分别相加？？？

### 对Series的values进行正弦变换

```python
print(np.sin(series_custom))

# 求series的最大值
print(np.max(series_custom))
```

### 数据中寻找指定范围的数值

#### 寻找评分>50的数据

```python
series_greater_than_50 = series_custom[series_custom > 50]
print(series_greater_than_50)
```

#### 寻找评分在(50, 75)的数据

```python
criteria_one = series_custom > 50
criteria_two = series_custom < 75
both_criteria = series_custom[criteria_one & criteria_two]
print(both_criteria)
```

### 对Series数据进行求均值
同维度的数据进行求均值（e.g.不同的评分系统求均值）

```python
rt_critics = Series(fandango['RottenTomatoes'].values, index=fandango['FILM'])
rt_users = Series(fandango['RottenTomatoes_User'].values, index=fandango['FILM'])
rt_mean = (rt_critics + rt_users)/2

print(rt_mean)
```
